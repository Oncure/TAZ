```python
import numpy as np
import matplotlib.pyplot as plt


X = np.linspace(0,4,1000)
Y = np.pi/2 * X * np.exp(-np.pi/4 * X*X)

plt.plot(X, Y, '-k', label='Wigner')
plt.plot(X, np.exp(-X), '--b', label='Poisson')
plt.xlabel('Normalized level-density', fontsize=16)
plt.ylabel('Probability Density', fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
X = np.linspace(0,3,1000)
Y = np.exp(-X/2) / np.sqrt(2*np.pi * X)

plt.plot(X, Y, '-k')
plt.xlabel('Normalized Neutron Widths', fontsize=16)
plt.ylabel('Probability Density', fontsize=16)
plt.xlim(0,2.5)
plt.ylim(0,4)
plt.grid(True)
plt.show()
```
