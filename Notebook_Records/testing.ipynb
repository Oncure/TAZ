```python
import numpy as np
import matplotlib.pyplot as plt

def FileReader(file):
    with open(file, 'r') as contents:
        txt = contents.read()
        CSD = txt.split('\n')
        E   = np.array([float(row.split()[0]) for row in CSD])
        CS  = np.array([float(row.split()[1]) for row in CSD])
    return E, CS

plt.figure()
E, CS = FileReader('/Users/colefritsch/ENCORE/Python_ENCORE/Notebook_Records/DT.txt')
idx = np.argmax(CS)
print(f'DT   = ({E[idx]:.4e}, {CS[idx]:.4e})')
plt.loglog(0.6*E, CS, '-k', label='D-D')
E, CS = FileReader('/Users/colefritsch/ENCORE/Python_ENCORE/Notebook_Records/HeHe.txt')
idx = np.argmax(CS)
print(f'HeHe = ({E[idx]:.4e}, {CS[idx]:.4e})')
plt.loglog(E, CS, ':b', label='3He-3He')
E, CS = FileReader('/Users/colefritsch/ENCORE/Python_ENCORE/Notebook_Records/p-11B.txt')
idx = np.argmax(CS)
print(f'pB   = ({E[idx]:.4e}, {CS[idx]:.4e})')
plt.loglog(700/599*E, CS, '-.r', label='p-11B')
E, CS = FileReader('/Users/colefritsch/ENCORE/Python_ENCORE/Notebook_Records/D-He.txt')
idx = np.argmax(CS)
print(f'DHe  = ({E[idx]:.4e}, {CS[idx]:.4e})')
plt.loglog(E*280/413, CS, '--g', label='D-3He')

plt.xlim(1e-2, 3e1)
plt.ylim(1e-2,2e1)
plt.xlabel('Energy (MeV)')
plt.ylabel('Cross-section (barns)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
